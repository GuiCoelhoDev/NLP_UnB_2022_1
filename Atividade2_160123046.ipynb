# Reading CSV

```python

pd.set_option('max_colwidth', None)
df = pd.read_csv('https://raw.githubusercontent.com/viniciusrpb/cic0269_natural_language_processing/main/corpus_tweets/twitter-2013train-A.txt', sep='\t', header=None, encoding='utf-8')
df.columns = ["id", "polarity", "text"]
df
```

# Removing duplicates (Removeu 35 tweets)

```python
df.drop_duplicates(subset='text',inplace=True)
df= df.reset_index(drop=True)
df
```

# Remoção de stopwords e subsituição de emojis em palavras

```python
stop_list_nltk = nltk.corpus.stopwords.words('english')
rt_words = ['rt', 'rts', 'retweet', '']
emoji_dictionary = {':\)': 'happy', ':(': "sad", ':|': "neutral"}

stop_words = set().union(stop_list_nltk, rt_words)

stop_words
```

```python
df['without_stopwords_and_emojis'] = df['text'].str.lower()\
.str.replace('(http\S+)', ' ')\
.apply(lambda x: [token for token in re.split(r'[\s!$@#,.;]+',x) if not token in stop_words])\
.apply(lambda x: [token.replace(':)', 'happy') for token in x ])\
.apply(lambda x: [token.replace(':(', 'sad') for token in x ])\
.apply(lambda x: [token.replace(':|', 'neutral') for token in x ])

df

```

# Stemmização

```python
from nltk.stem.porter import *
stemmer = PorterStemmer()
```

```python
df['final_processing_with_stemming'] = df['without_stopwords_and_emojis'].apply(lambda rows: [stemmer.stem(row) for row in rows if row != ''])
df
```

# Vocabulário do corpus

```python
corpus_tokens = []

for i in range(len(df['final_processing_with_stemming'])):
  for token in df['final_processing_with_stemming'][i]:
    corpus_tokens.append(token)


print("Length of list of tokens:", len(corpus_tokens))
print(corpus_tokens)

corpus_vocab = list(dict.fromkeys(corpus_vocab))
print("Length of list of vocabulary (without duplicated):", len(corpus_vocab))
print(corpus_vocab)

for word in corpus_vocab:
  match = re.findall(r'\:\)', word)
  if match != []:
    print(match)



  
```
